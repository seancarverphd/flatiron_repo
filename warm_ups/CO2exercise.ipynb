```python
def tr_te_split(X, Y, cond, notcond):
    nc = list(notcond[0])
    c = list(cond[0])
    trX = X.iloc[nc,]
    teX = X.iloc[c,]
    trY = Y[nc]
    teY = Y[c]
    return trX, trY, teX, teY
```
